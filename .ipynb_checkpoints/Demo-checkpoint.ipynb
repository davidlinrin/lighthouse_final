```python
import requests,json
import pandas as pd
```

```python
URL = "http://127.0.0.1:5000/"
```

# 1) Recommendation for new users with no user data

#### FINAL FANTASY XIV Online, Raft, Counter-Strike: Global Offensive,Dota 2, Monster Hunter: World, Terraria

```python
endpoint = "games_alike"
r = requests.post(url = URL+endpoint, json = {'game':"FINAL FANTASY XIV Online", 'top':4})
r.json()
```

# 2) Recommendation for users on the platform

# 3) Rating prediction for specific user and application

#### USER: 76561197960274289, 76561197960274467, 76561198057398497
#### GAMES: Monster Hunter: World: 582010, Terraria: 105600, Raft: 648800, CSGO :730, Dota 2: 570

# 4) Database requests

### Game id and name conversion
### Monster Hunter: World: 582010, Terraria: 105600, Raft: 648800, CSGO :730, Dota 2: 570

```python
endpoint = "id_to_game"
r = requests.post(url = URL+endpoint, json = {'game_ID':582010})
r.json()
```

```python
endpoint = "game_to_id"
r = requests.post(url = URL+endpoint, json = {'game_name':'Terraria'})
r.json()
```
