```python
from scipy.sparse import csr_matrix
from sklearn.metrics.pairwise import linear_kernel
```

```python
games_index = copy.deepcopy(games)
games_index.index = games_index['steam_appid']
games_index.drop(['name','steam_appid'], axis =1, inplace = True)
games_index['controller_support'] = np.where(games_index['controller_support']=='None',0,1)
games_index['is_free'] = np.where(games_index['is_free']==False,0,1)
games_index['required_age'] = np.where(games_index['required_age']>0,1,0)
games_index['type'] = np.where(games_index['type']=='game',1,
                              np.where(games_index['type']=='mod',2,
                                      np.where(games_index['type']=='dlc',3,0)))
games_index['windows'] = np.where(games_index['windows']==True,1,0)
games_index['mac'] = np.where(games_index['mac']==True,1,0)
games_index['linux'] = np.where(games_index['linux']==True,1,0)
games_index['price_overview'] = games_index['price_overview'].fillna(0)
games_index
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
games_index_matrix = csr_matrix(games_index.values)
```

```python
similarities = cosine_similarity(games_index_matrix)
```

```python
indices = pd.Series(games_index.reset_index().index, index=games['name']).drop_duplicates()
```

```python
indices
```

```python
def get_recommendations(title, cosine_sim=similarities, top = 3):
    # Get the index of the movie that matches the title
    idx = indices[title]

    # Get the pairwsie similarity scores of all movies with that movie
    sim_scores = list(enumerate(similarities[idx]))

    # Sort the movies based on the similarity scores
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)

    # Get the scores of the 10 most similar movies
    sim_scores = sim_scores[1:top+1]

    # Get the movie indices
    game_indices = [i[0] for i in sim_scores]
    
    # Return the top 10 most similar movies
    return games['name'].iloc[game_indices]
```

```python
get_recommendations('Counter-Strike: Global Offensive', top = 5)
```
