```python
class Solution(object):
    def removeDuplicateLetters(self, s):
        stack1 = []
        stack2 = []
        a = s
        for i in a:
            if i in stack1:
                test = []
                for temp in stack1:
                    test.append(temp)
                test.remove(i)
                test.append(i)
                if "".join(test) < "".join(stack1):
                    stack1 = test
                else:
                    next
            else:
                stack1.append(i)


        for i in a[::-1]:
            if len(stack2)==len(set(a)):
                break
            if i in stack2:
                test = []
                for temp in stack2:
                    test.append(temp)
                test.remove(i)
                test.append(i)
                if "".join(test) < "".join(stack2):
                    stack = test
                else:
                    next
            else:
                stack2.append(i)



        if "".join(stack1) < "".join(stack2[::-1]):
            final = "".join(stack1)
        else:
            final = "".join(stack2[::-1])
            
        return(final)
```
